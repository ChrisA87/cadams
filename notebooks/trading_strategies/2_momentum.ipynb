# Strategies Based on Momentum

There are 2 basic types of momentum strategies:

1) **Cross-sectional** <br>
Where you would buy an asset if it has - over a certain period of time - outperfermed relative to its peers or some other industry benchmark, and sell an asset if it has underperformed. 

2) **Time series**<br>
Where you would buy an asset that has recently performed well, benchmarked against _itself_ over some period of time and vice versa.


This notebook focuses on the implementation of backtesting time series momentum strategies.

## Reading the data

```python
raw = pd.read_csv('http://hilpisch.com/pyalgo_eikon_eod_data.csv', index_col=0, parse_dates=True).dropna()
raw
```

## Set up basic position and returns

Using the price of gold in USD, in this basic example, the position is **+1** when the return of the previous day was positive and **-1** when negative.

```python
plt.style.use('seaborn')

data = raw[['XAU=']].rename(columns={'XAU=': 'price'})
data['returns'] = np.log(data['price'] / data['price'].shift())
data['position'] = np.sign(data['returns'])
data['strategy'] = data['position'].shift() * data['returns']

(data[['returns', 'strategy']].dropna()
 .cumsum()
 .apply(np.exp)
 .plot(figsize=(10, 6),
       title='Gross performance of gold price (USD) vs momentum (last day only)'))
```

Here, the strategy's return significantly underperforms the base return.

This strategy can be generalised to longer time windows by calculating the rolling mean.

## Generalising over the last 3 days returns

```python
def plot_momentum(returns: pd.Series, n_periods: int=1):
    data = returns.rename('returns').to_frame()
    data['position'] = np.sign(data['returns'].rolling(n_periods).mean())
    data['strategy'] = data['position'].shift() * data['returns']

    (data[['returns', 'strategy']].dropna()
     .cumsum()
     .apply(np.exp)
     .plot(figsize=(10, 6),
           title=f'Gross performance of gold price (USD) vs momentum - last {n_periods} day(s)'))
```

```python
plot_momentum(data['returns'], 3)
```

Over 3 days, the strategy significantly outperforms the base return.

It's important to understand however, that this strategy is very sensitive to the time window parameter.
For example, over a rolling 2 days also underperforms the base, whilst 4 outperforms again.

```python
plot_momentum(data['returns'], 2)
```

```python
plot_momentum(data['returns'], 4)
```

## Comparing multiple strategies

```python
for n in range(1, 10, 2):
    data[f'position_{n}'] = np.sign(data['returns'].rolling(n).mean())
    data[f'strategy_{n}'] = data['returns'] * data[f'position_{n}'].shift()

fig = plt.figure(figsize=(12, 8))
ax = plt.gca()

(data.dropna()
 .loc[:, 'returns']
 .cumsum()
 .apply(np.exp)
 .plot(ax=ax, legend=True,
       title='Volitility of time series momentum strategy with different time windows.'))
(data.filter(regex='strategy_')
 .dropna()
 .cumsum()
 .apply(np.exp)
 .plot(ax=ax, style='--', alpha=0.4))
```
